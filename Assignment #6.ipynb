```python
words = [] 
for s in sent: 
    for w in word_tokenize(s): 
        words.append(w)
```

```python
extrastop = ['``',"''","'re","'s","'re",'``',"''","'ll","--","\'\'","...",
             "n\'t",'one','would','use','subject','from',"\'m","\'ve", "’","HAMLET","And", "The", "That", "But", "To", "O", "HORATIO"
            "KING", "POLONIUS", "A", "What", "HORATIO"]
```

```python
# I added some of these words because they are for those that have the most lines since script is set up like this: 
# HAMLET : "dialgoue here..." so I thought it was important to get rid of these words as they do not give an accurate 
#representation of the top 20 words. 
```

```python
myStopWords = list(punctuation) + stopwords.words('english') + extrastop
```

```python
myStopWords
```

```python
wordsNoStop = []
for i in words:
    if i not in myStopWords:
        wordsNoStop.append(i)
#print(words)
print(wordsNoStop)
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(wordsNoStop)
```

```python
freq
```

```python
for i in sorted(freq, key=freq.get, reverse=True)[:20]:
    print(i,freq[i])
```

```python
# I think that these words do give an accurate representation of the play as Hamlet is talking about himself a lot of the play
# and in most Shakespeare's plays, the word 'I' is very vital. The rest of the words are common words that Shakespeare
# uses in his plays. 
```

```python
[w for w in word_tokenize(ListMovieReviews[0].lower()) if w not in myStopWords]
```

```python
listOfMovieWords = []
for i in ListMovieReviews:
    listOfMovieWords.append([w for w in word_tokenize(i.lower()) if w not in myStopWords])
```

```python
listOfMovieWords[0]
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
listOfStemmedWords = []
for i in listOfMovieWords:
    listOfStemmedWords.append([p_stemmer.stem(w) for w in i])
```

```python
listOfStemmedWords[0]
```

```python
!pip install gensim
```

```python
from gensim import corpora, models
import gensim
```

```python
dictionary = corpora.Dictionary(listOfStemmedWords)
```

```python
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in listOfStemmedWords]
```

```python
print(corpus[1])
```

```python
print(dictionary.token2id['best'])
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics=2, 
                                           id2word = dictionary, 
                                           passes=20)
```

```python
for i in ldamodel.print_topics(num_topics=2, num_words=20):
    print(i)
```

I see a mixed split in the two categories. I see the word "baseball" in one of the categories and "funni" in the other. The word "zombi" is also with the category that has the word "comedi". It did not perfectly split comedy and drama. I find that it picked words that were in the comedy reviews but did far less of a good job with picking the words that were in the drama reviews. Overall, when spliting between drama and comedy, it did not do a good job but it did a better job in picking words from the comedy reviews.
