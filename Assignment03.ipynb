# Creating Graphs to Compare Violent Crime Alabama vs. Oregon

```python
y=df_a.plot(kind='barh',figsize=(12,6))
y.set_title('Alabama Violent Crime')
y.legend(['x-axis shows value >500'])
```

```python
x=df_o.plot(kind='barh',figsize=(12,6))
x.set_title('Oregon Violent Crime')
x.legend(['x-axis shows value >300'])
```

It may seem that Oregon has more overall crime than Alabama but the X-axis value is important to pay attnetion to. Oregon's X-axis value is less than Alabama's X-axis value.
